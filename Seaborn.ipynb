JointPlot
""" Its is combination of both Scatter Plot & Histogram """

```python
import seaborn as sns
import matplotlib.pyplot as plt

Data = pd.read_csv('Data.csv')

sns.jointplot(x='Sepal_Length',y='Petal_Length',data=Data)
plt.show()
```
