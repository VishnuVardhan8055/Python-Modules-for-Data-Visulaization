Python Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,11)
x
```

```python
y = 2*x
y
```

```python
# Line Plot
plt.plot(x,y,color="green",linewidth=1)
plt.title("Lie Plot",color="Red")
plt.xlabel("Ages")
plt.ylabel("Years")
plt.show()
```

```python
# Line Plot For two Lines
y2 = 3*x

plt.plot(x,y,color="red",linestyle=":",linewidth=3,label="White")
plt.plot(y2,y,color="black",linestyle="-",linewidth=1,label="RED")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line_Plot")
plt.legend()
plt.show()
```

```python
# Sub Plots

plt.subplot(1,2,1)
plt.plot(x,y,color="red",linestyle=":",label="data")
plt.legend()

plt.subplot(1,2,2)
plt.plot(y,y2,color="green",linestyle="-",label="Content")
plt.legend()

plt.show()
```

Bar Plot

```python
import matplotlib.pyplot as plt
Market_Price ={'Mangos':120,'Oranges':160,'Kiwi':210,'Bananas':75,'Grapes':190}
Fruits=list(Market_Price.keys())
Prices=list(Market_Price.values())
color=['yellow','orange','green','yellow','black']
plt.bar(Fruits,Prices,color=color,width=0.4)
plt.xlabel('Fruits',color='Blue')
plt.ylabel('Prices/kg',color='Blue',)
plt.title("Prices Of Fruits")
plt.grid(True)
plt.show()
```

Scatter Plot

```python
import matplotlib.pyplot as plt
x=[12,23,45,65,67,87,11,45,56]
a=[10,23,40,56,76,78,10,54,65]
C=['green','gold','red','blue','black','orange','red','brown','green']
plt.scatter(x,a,marker="^",color=C,s=30)
plt.show()
```

```python
# Adding two Scatter Plot
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]
y1 = [12,34,56,78,90,45,67]
y2 = [2,45,43,60,87,29,70]

plt.scatter(x,y1,marker="*",color='green',s=25,label="Growth")
plt.xlabel("Market Status")
plt.ylabel("Market Vale",color="Blue")
plt.scatter(x,y2,marker="^",color='red',s=32,label="Fall")
plt.legend()
plt.title("Market Status")
plt.show()
```

```python
# Adding Scatter Sub Plots
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]
y1 = [12,34,56,78,90,45,67]
y2 = [2,45,43,60,87,29,70]

plt.subplot(1,2,1)
plt.scatter(x,y1,marker="*",color='green',s=25,label="Growth")
plt.xlabel("Market Status")
plt.ylabel("Market Vale",color="Blue")
plt.legend()

plt.subplot(1,2,2)
plt.scatter(x,y2,marker="^",color='red',s=32,label="Fall")
plt.xlabel("Market Status")
plt.ylabel("Market Vale",color="Blue")
plt.legend()
plt.title("Market Status")
plt.show()
```

BOX Plot

```python
# Creating Box Plot
import matplotlib.pyplot as plt
x = [1,4,7,8,3]
y = [13,23,43,56,78]
data=list([x,y])
plt.boxplot(data)
plt.show
```

Violin Plot

```python
import matplotlib.pyplot as plt
x = [1,4,7,8,3]
y = [13,23,43,56,78]
data=list([x,y])
plt.violinplot(data,showmedians=True)
plt.show
```

Pie Chart

```python
import matplotlib.pyplot as plt
x=[20,30,45,15]
y=["house","food","Education","savings"]
plt.pie(x,labels=y,radius=0.8)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[20,30,45,15]
y=["house","food","Education","savings"]
plt.pie(x,labels=y,radius=0.8,autopct='%0.1f%%')# """ Here autopct is to calculate the percentage of each values , and 1f is to show the entire value os float. """
plt.show()
```

DoughNut--Chart

```python
import matplotlib.pyplot as plt
x=[20,30,45,15]
y=["house","food","Education","savings"]

plt.pie(x,labels=y,radius=1,autopct='%0.1f%%')
plt.pie([1],colors=['w'],radius=0.5)
plt.show()
```
